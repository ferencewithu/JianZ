# UML 类图分析

## Project（项目模型聚合根）
**功能**：管理整个工程，包含音频项、顺序和剪辑参数，负责工程文件的保存/加载。

- **属性**
  - `- items: List<AudioItem>` → 保存所有音频项，外部不直接操作
  - `- sequence: List<int>` → 保存播放顺序
  - `- clips: Dict<ItemID, ClipConfig>` → 保存每个音频项的剪辑参数

- **方法**
  - `+ loadFolder(path)` → 导入文件夹，生成 AudioItem
  - `+ shuffle()` → 打乱顺序
  - `+ reorder(newOrder)` → 调整顺序
  - `+ getConfig(itemID)` → 获取某个音频项的剪辑参数
  - `+ save(path)` → 保存工程文件
  - `+ load(path)` → 加载工程文件

---

## AudioItem（音频文件元数据）
**功能**：描述单个音频文件的基本信息，保持只读。

- **属性**
  - `+ id: str` → 唯一标识
  - `+ path: str` → 文件路径
  - `+ durationMs: float` → 时长
  - `+ sampleRate: int` → 采样率
  - `+ channels: int` → 声道数
  - `- meta: dict` → 额外元数据，内部使用

- **方法**
  - `+ fromPath(path)` → 工厂方法，生成 AudioItem
  - `+ toDict()` → 序列化为字典

---

## ClipConfig（剪辑参数配置）
**功能**：保存剪辑参数，支持校验与序列化。

- **属性**
  - `+ startMs: float` → 起始时间
  - `+ endMs: float` → 结束时间
  - `+ fadeInMs: float` → 淡入时长
  - `+ fadeOutMs: float` → 淡出时长
  - `+ gainDb: float` → 增益

- **方法**
  - `+ validate()` → 检查参数合法性
  - `+ toDict()` → 序列化
  - `+ fromDict()` → 反序列化

---

## AudioProcessor（音频处理引擎）
**功能**：执行剪辑、拼接、合成等核心处理。

- **方法**
  - `+ applyClip(item: AudioItem, cfg: ClipConfig) -> TempAudio` → 对单个音频应用剪辑参数
  - `+ clipAndConcat(plan: Iterable[(AudioItem, ClipConfig)]) -> MixedAudio` → 批量处理并拼接成完整音频

---

## Exporter（导出器）
**功能**：将合成结果编码并写入文件。

- **方法**
  - `+ exportAndSave(data: MixedAudio, targetPath: str, format: str)` → 导出成品音频

---

## PlayerService（播放服务）
**功能**：负责预览播放，控制播放状态。

- **方法**
  - `+ play(source)` → 播放音频
  - `+ pause()` → 暂停播放
  - `+ stop()` → 停止播放
  - `+ seek(ms)` → 跳转播放位置

---

## MainWindow（UI 控制器）
**功能**：协调用户操作与后台逻辑。

- **方法**
  - `+ openImport()` → 打开导入对话框
  - `+ refreshPlaylist()` → 刷新播放列表
  - `+ onPreview()` → 触发预览
  - `+ onExport()` → 触发导出

```python
# 9 分段函数
# 32502420—————樊陈轩宇————fromZtoU
def func(x: int) -> int: #樊陈轩宇_fromZtoU 定义分段函数
    y = None # 樊陈轩宇_fromZtoU 初始化 y
    if x < 0: # 樊陈轩宇_fromZtoU 判断 x 小于 0
        y = 0   # 樊陈轩宇_fromZtoU y 赋值为 0
    elif x < 5: # 樊陈轩宇_fromZtoU 判断 x 小于 5
        y = x # 樊陈轩宇_fromZtoU y 赋值为 x
    elif x < 10: # 樊陈轩宇_fromZtoU 判断 x 小于 10
        y = 3*x-5 # 樊陈轩宇_fromZtoU y 赋值为 3x-5
    elif x < 20: # 樊陈轩宇_fromZtoU 判断 x 小于 20
        y = 0.5*x-2 # 樊陈轩宇_fromZtoU y 赋值为 0.5x-2
    else: # 樊陈轩宇_fromZtoU 判断 x 大于等于 20
        y = 0 # 樊陈轩宇_fromZtoU y 赋值为 0
    return y # 樊陈轩宇_fromZtoU 返回 y

arr = [-2, 3, 7, 15, 25] # 樊陈轩宇_fromZtoU 测试数据
for i in arr: # 樊陈轩宇_fromZtoU 遍历测试数据
    print(func(i)) # 樊陈轩宇_fromZtoU 输出结果
```

```python
# 10 模拟蒙提霍尔问题
import random

def game_body():
    arr = ['goat','goat','car']
    random.shuffle(arr)
    firstKey = int(input('Choose a door to open (1~3):'))
    goatKey = arr.index('goat')
    print(f'"goat" behind the door {goatKey+1}')
    choice = input('Switch ur door'+'?(y/n)')
    if choice == 'y':
        print('You Win!') if arr[4 - firstKey - goatKey] == 'car' else print('You Lose!')
    else:
        print('You Win!') if arr[firstKey-1] == 'car' else print('You Lose!')

game_body()
```
